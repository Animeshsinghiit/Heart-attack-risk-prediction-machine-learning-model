**Let's load the data set**

```python
heart=pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv")
o2sat=pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/o2Saturation.csv")
```

**Now, Let's see top 5 rows to get an idea about all the features.**

```python
heart.head()
```

```python
o2sat.head()
```

***Let me introduce all the columns to you and there meanings***

   - **Age :** Age of the patient
   - **Sex :** Sex of the patient
   - **exang:** exercise induced angina (1 = yes; 0 = no)
   - **ca:** number of major vessels (0-3)
   - **cp :** Chest Pain type chest pain type
        Value 1: typical angina
        Value 2: atypical angina
        Value 3: non-anginal pain
        Value 4: asymptomatic
   - **trtbps :** resting blood pressure (in mm Hg)
   - **chol :** cholestoral in mg/dl fetched via BMI sensor
   - **fbs :** (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
   - **rest_ecg :** resting electrocardiographic results
        Value 0: normal
        Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
        Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
   - **thalach :** maximum heart rate achieved
   - **target :** 0= less chance of heart attack 1= more chance of heart attack

**In this notebook we will work with heart dataset**
> To do so let's seperate the feature set from the target, in other words let's seperate the features and labels

```python
data=heart.iloc[:,:-1]
label=heart.iloc[:,-1]
```

**Now let's describe our data and see some statistics**

```python
data.describe()
```

As we can observe from min,max,mean that all the columns are at different scale. So we will Standardize our data as
> Standardization is useful when your data has varying scales and the algorithm you are using does make assumptions about your data having a Gaussian distribution, such as linear regression, logistic regression, and linear discriminant analysis.

we will use Standard Scaler for it

```python
from sklearn.preprocessing import StandardScaler
```

```python
#let's take all the feature names(column names) into one list.
features=data.columns
ss=StandardScaler()
data_tf=ss.fit_transform(data)
data_tf=pd.DataFrame(data_tf,columns=features)
data_tf.head()
```

Let's describe our data again

```python
data_tf.describe()
```

Now if we carefully observe our data we will observe that our data is now at same scale

And let's do some analysis on our target columns.

```python
label.value_counts()
```

```python
print('% of 1 label in data set: {ones} %'.format(ones=165*100/(165+138)))
print('% of 0 label in data set: {ones} %'.format(ones=138*100/(165+138)))
```

we can observe that our data is not much imbalanced as there are **165 datapoints with 1** (54.4 %)label and **138 with 0 label** (45.5%)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(n_jobs=-1,verbose=3,warm_start=True)
param_grid={
    'solver':['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'max_iter':[150,200,100],
}
gs=GridSearchCV(lr,param_grid=param_grid,refit=True,n_jobs=-1,cv=5,scoring='f1',verbose=3,error_score='raise',return_train_score=True)
gs.fit(data_tf,label)
```

```python
gs.cv_results_
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

```python
gs.scorer_
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_jobs=-1,verbose=3)
param_grid={
    'bootstrap':[True,False],
    'class_weight':["balanced", "balanced_subsample"],
    'n_estimators':[100,150,200,250,300,350,400]
}
gs2=GridSearchCV(rfc,param_grid=param_grid,refit=True,n_jobs=-1,cv=5,scoring='f1',verbose=3,error_score='raise',return_train_score=True)
gs2.fit(data_tf,label)
```

```python
gs2.best_estimator_
```

```python
gs2.best_score_
```

**We can see here that our LogisticRegression worked better then RandomForestClassifier
